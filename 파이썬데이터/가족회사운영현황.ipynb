```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
path = 'C:/windows/fonts/H2GTRM.TTF'
font_name = fm.FontProperties(fname = path).get_name()
plt.rc('font',family=font_name)
f = open('data/.csv')
data = csv.reader(f)

header = next(data)
data1= []
xlabel= []
next(data)
# print(header)
xlabel.append(header[4])
xlabel.append(header[7])
xlabel.append(header[10])
xlabel.append(header[13])
xlabel.append(header[16])
xlabel.append(header[19])

# print(xlabel)
data2=[]
for row in data:
#       
    if row[0] =="":
        row[3:]=map(int,row[3:]);
        if row[2] != "":

            d= row[4],row[7],row[10],row[13],row[16],row[19]
            data2.append(row[2])
            data1.append(d)


            


# 
plt.figure(figsize=(20,10))
for i in range(len(data2)):
    plt.plot(xlabel,data1[i], label=data2[i])
plt.legend()
plt.show()


# 
for i in range(len(data2)):
    plt.figure(figsize = (20,10))
    plt.title(data2[i])
    plt.bar(xlabel,data1[i])
    plt.show()

# plt.figure(figsize = (20,10))
# plt.title(data2[0])
# plt.bar(xlabel,data1[0])
# plt.show()
        
```
